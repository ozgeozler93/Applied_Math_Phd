**Polynomial Regression**

```python
import pandas as pd

df = pd.read_csv(r"../Course_Data/polynomial_regression.csv", sep=';')


df.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(df['x'])
y = np.array(df['y'])

plt.scatter(x, y, color="red")

plt.show()
```

```python
plt.plot(x, y, 'o')

m, b = np.polyfit(x, y, 1)

x_ = np.linspace(0,2,50)
y_ = b * np.ones_like(x_) + m * x_

plt.plot(x_, y_)
```

```python
plt.plot(x, y, 'o')

a, b, c = np.polyfit(x, y, 2)

x_2 = np.linspace(0,2,50)
y_2 = []

y_2 = a * x_2**2 + b * x_2 + c * np.ones_like(x_2)

plt.plot(x_2, y_2)
```

```python
print(a,b,c)
```

```python
plt.plot(x, y, 'o')

a, b, c, d, e, f = np.polyfit(x, y, 5)

x_ = np.linspace(0,2,50)

y_ = a * x_**5 + b * x_**4 + c * x_**3 + d * x_**2 + e * x_ + f * np.ones_like(x_)

plt.plot(x_, y_)
```

```python
fig= plt.figure()

axes= fig.add_axes([0,0,0.8,0.8])
axes.plot(x, y, 'o')

degree = 10

coef = np.polyfit(x, y, degree)

x_ = np.linspace(0, 2, 50)
y_ = coef[degree]* np.ones_like(x_)

for i in range(degree):
    y_ += coef[i] * x_**[degree - i]

axes.plot(x_, y_)

axes.set_xlim([0,2])

axes.set_ylim([-0.3,4])

plt.show()

x_10, y_10 = x_, y_
```

```python
fig= plt.figure()

axes= fig.add_axes([0,0,0.8,0.8])
axes.plot(x, y, 'o')

degree = 15

coef = np.polyfit(x, y, degree)

x_ = np.linspace(0, 2, 50)
y_ = coef[degree]* np.ones_like(x_)

for i in range(degree):
    y_ += coef[i] * x_**[degree - i]

axes.plot(x_, y_)

axes.set_xlim([0,2])

axes.set_ylim([-0.3,4])

plt.show()

x_15, y_15 = x_, y_
```

```python
# The data above is created with the formula y = x ** 2 + e, where e is some noise.
# Let's create new data with the same formula

def create_data(number_of_points):
    x = []
    y = []

    for i in range(number_of_points):
        x.append(np.random.uniform(low=0.0, high=2.0, size=None))
        y.append(x[-1]**2 + np.random.normal(loc=0.0, scale=0.25, size=None))
        
    return x,y
```

```python
x,y = create_data(20)

plt.scatter(x, y)
```

```python
x2,y2 = create_data(20)

fig= plt.figure()

axes= fig.add_axes([0.1,0.1,0.8,0.8])


axes.scatter(x2, y2)

axes.plot(x_2, y_2)

axes.plot(x_10, y_10)

#axes.plot(x_15, y_15)


axes.set_xlim([0,2])

axes.set_ylim([-0.3,4])

plt.show()
```
